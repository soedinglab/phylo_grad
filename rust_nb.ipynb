```rust
pub fn create_example_binary_tree() -> BinaryTreeNode<usize> {
    BinaryTreeNode {
        value: 0,
        left: Some(Box::new(BinaryTreeNode {
            value: 1,
            left: Some(Box::new(BinaryTreeNode {
                value: 3,
                left: None,
                right: None,
            })),
            right: Some(Box::new(BinaryTreeNode {
                value: 4,
                left: None,
                right: None,
            })),
        })),
        right: Some(Box::new(BinaryTreeNode {
            value: 2,
            left: Some(Box::new(BinaryTreeNode {
                value: 5,
                left: None,
                right: None,
            })),
            right: Some(Box::new(BinaryTreeNode {
                value: 6,
                left: Some(Box::new(BinaryTreeNode {
                    value: 7,
                    left: None,
                    right: Some(Box::new(BinaryTreeNode {
                        value: 8,
                        left: Some(Box::new(BinaryTreeNode {
                            value: 9,
                            left: None,
                            right: Some(Box::new(BinaryTreeNode {
                                value: 10,
                                left: None,
                                right: None,
                            })),
                        })),
                        right: None,
                    })),
                })),
                right: None,
            })),
        })),
    }
}
```

```rust
let root = create_example_binary_tree();
```

```rust
let result = 
	root.dfs_postorder_iter()
		.map(|val| val.to_string())
		.collect::<Vec<String>>()
		.join(", ");
```

```rust
// result: 3, 4, 1, 5, 10, 9, 8, 7, 6, 2, 0
println!("{result}");
```

```rust
#[derive(Debug)]
struct TreeFixed<Data, const TREE_SIZE: usize> {
    data: [Data; TREE_SIZE],
}

/* Implement as a trait to have common interface? */
type Id = usize;
impl<const TREE_SIZE: Id, const TOTAL: usize> TreeFixed<FelsensteinNodeStd<TOTAL>, TREE_SIZE> {
    fn left(id: Id) -> Option<Id> {
        let left_id = id * 2 + 1;
        if left_id < TREE_SIZE {
            return Some(left_id);
        } else {
            return None;
        }
    }
    fn right(id: Id) -> Option<Id> {
        let right_id = id * 2 + 2;
        if right_id < TREE_SIZE {
            return Some(right_id);
        } else {
            return None;
        }
    }
    fn trivial() -> Self {
        let data = [FelsensteinNodeStd::<TOTAL>::trivial(); TREE_SIZE];
        Self{data: data}
    }
}

#[derive(Debug, Clone, Copy)]
pub struct FelsensteinNodeStd<const TOTAL: usize> {
    log_p: [Float; TOTAL],
    distance: Float,
}
impl<const TOTAL: usize> FelsensteinNodeStd<TOTAL> {
    fn trivial() -> Self {
        FelsensteinNodeStd{
            log_p: [1 as Float; TOTAL],
            distance: 1 as Float,
        }
    }
}

let tree: TreeFixed::<FelsensteinNodeStd::<5>, {1<<5}>= TreeFixed::<FelsensteinNodeStd::<5>, {1<<5}>::trivial();

tree.data
```
