```rust
pub fn create_example_binary_tree() -> BinaryTreeNode<usize> {
    BinaryTreeNode {
        value: 0,
        left: Some(Box::new(BinaryTreeNode {
            value: 1,
            left: Some(Box::new(BinaryTreeNode {
                value: 3,
                left: None,
                right: None,
            })),
            right: Some(Box::new(BinaryTreeNode {
                value: 4,
                left: None,
                right: None,
            })),
        })),
        right: Some(Box::new(BinaryTreeNode {
            value: 2,
            left: Some(Box::new(BinaryTreeNode {
                value: 5,
                left: None,
                right: None,
            })),
            right: Some(Box::new(BinaryTreeNode {
                value: 6,
                left: Some(Box::new(BinaryTreeNode {
                    value: 7,
                    left: None,
                    right: Some(Box::new(BinaryTreeNode {
                        value: 8,
                        left: Some(Box::new(BinaryTreeNode {
                            value: 9,
                            left: None,
                            right: Some(Box::new(BinaryTreeNode {
                                value: 10,
                                left: None,
                                right: None,
                            })),
                        })),
                        right: None,
                    })),
                })),
                right: None,
            })),
        })),
    }
}
```

```rust
#[derive(Debug)]
struct TreeFixed<Data, const TREE_SIZE: usize> {
    data: [Data; TREE_SIZE],
}

/* Implement as a trait to have common interface? */
type Id = usize;
impl<const TREE_SIZE: Id, const TOTAL: usize> TreeFixed<FelsensteinNodeStd<TOTAL>, TREE_SIZE> {
    fn left(id: Id) -> Option<Id> {
        let left_id = id * 2 + 1;
        if left_id < TREE_SIZE {
            return Some(left_id);
        } else {
            return None;
        }
    }
    fn right(id: Id) -> Option<Id> {
        let right_id = id * 2 + 2;
        if right_id < TREE_SIZE {
            return Some(right_id);
        } else {
            return None;
        }
    }
    fn trivial() -> Self {
        let data = [FelsensteinNodeStd::<TOTAL>::trivial(); TREE_SIZE];
        Self{data: data}
    }
}

#[derive(Debug, Clone, Copy)]
pub struct FelsensteinNodeStd<const TOTAL: usize> {
    log_p: [Float; TOTAL],
    distance: Float,
}
impl<const TOTAL: usize> FelsensteinNodeStd<TOTAL> {
    fn trivial() -> Self {
        FelsensteinNodeStd{
            log_p: [1 as Float; TOTAL],
            distance: 1 as Float,
        }
    }
}

let tree: TreeFixed::<FelsensteinNodeStd::<5>, {1<<5}>= TreeFixed::<FelsensteinNodeStd::<5>, {1<<5}>::trivial();

tree.data
```

```rust
//tutorial-read-01.rs
use std::{
    env,
    error::Error,
    fs::File,
    path::Path,
    process,
    io::self,
    io::BufReader,
    io::BufRead,
};


pub fn read_tree<P>(filename: P) -> Result<(), Box<dyn Error>>
where
    P: AsRef<Path>,
{
    let file = File::open(filename)?;
    let mut bufreader = BufReader::new(file);
    
    let mut s = String::new();
    bufreader.read_line(&mut s);
    println!("{s}");
    let sequence_length = s.trim().parse::<usize>()?;
    
    let mut rdr = csv::Reader::from_reader(bufreader);
    let mut i = 0i32;
    for result in rdr.records() {
        let record = result?;
        println!("{:?}", record);
        i+=1;
        if(i >= 10) {
            break;
        }
    }
    Ok(())
}

if let Err(err) = read_tree("/home/jora/src/rust/tree.txt") {
   println!("{}", err);
   process::exit(1);
}
```

```rust
pub fn read_csv<P>(
    filename: P,
) -> Result<Box<dyn Iterator<Item = Result<Record, csv::Error>>>, Box<dyn Error>>
where
    P: AsRef<Path>,
{
    let file = File::open(filename)?;
    let mut bufreader = BufReader::new(file);

    let mut s = String::new();
    bufreader.read_line(&mut s)?;
    let sequence_length = s.trim().parse::<usize>()?;
    //println!("{sequence_length}");

    let mut rdr = csv::Reader::from_reader(bufreader);
    Ok(rdr)
}
```
